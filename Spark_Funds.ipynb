# Checkpoint 1: Read data and inspect

Read the companies and rounds data into two data frames companies and rounds2 and inspect them.


```python
# Loading companies data from local drive , make permalink field lower to match with company_permalink of rounds2 data 

companies = pd.read_csv('D:\\Group-Assignments\\Course1\\companies.txt', sep='\t', encoding='ISO-8859-1')
# ignore the special characters
companies['permalink'] = companies[['permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
# change the permalink column values to lower case.
companies['permalink'] = companies['permalink'].str.lower()
companies.head()
```

```python
# inspect companies dataframe

print("*********************************  Info *******************************************") 
print(companies.info())
print("*********************************  Shape ******************************************") 
print(companies.shape)
print("**************************** Columns having null values *****************************")
print(companies.isnull().any())
print("*********************************  Description ************************************") 
companies.describe()
```

```python
# load rounds2 data 

rounds2 = pd.read_csv('D:\\Group-Assignments\\Course1\\rounds2.csv', encoding='ISO-8859-1')
# remove special characters from company_permalink column
rounds2['company_permalink'] = rounds2[['company_permalink']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
# change the company_permalink column values to lower case
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
rounds2.head()
```

```python
# inspect rounds2 dataframe

print("*********************************  Info *******************************************") 
print(rounds2.info())
print("*********************************  Shape ******************************************") 
print(rounds2.shape)
print("**************************** Columns having null values *****************************")
print(rounds2.isnull().any())
print("*********************************  Description ************************************") 
rounds2.describe()
```

```python
# How many unique companies are present in companies?

companies['permalink'].nunique()
```

## Checkpoint 2: Funding Type Analysis

` Calculate the average investment amount for each of the four funding types (venture, angel, seed, and private equity)`

```python
# Data Cleaning - Before we proceed further, we can remove the following columns, as they would not be used for any computation.
# funding_round_code, funded_at, home_page_url, status, state_code, region, city, founded_at

master_frame = master_frame.drop(['funding_round_permalink','funding_round_code', 'funded_at', 'homepage_url', 'status', 
                                  'state_code','region', 'city', 'founded_at'], axis=1) 
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# Remove all nan rows for column raised_amount_usd, as they would not add value to further analysis.

master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# Since the investment amount is to be determined in millions , convert the raised_amount_usd column from usd to millions

master_frame['raised_amount_usd'] = master_frame['raised_amount_usd'].apply(lambda x: (x/1000000))
master_frame.head()
```

```python
# Check if there are any rows that have the raised amount as 0 

len(master_frame.loc[(master_frame['raised_amount_usd'] == 0 )])
```

```python
# Since there are 415 rows with raised amount as 0, remove those

master_frame.drop(master_frame.loc[master_frame['raised_amount_usd'] == 0].index, inplace=True)
```

```python
# check if the rows have been removed successfully....

len(master_frame.loc[(master_frame['raised_amount_usd'] == 0 )])
```

```python
# Calculate the total of investment amount in each funding round type

df_funding_type = master_frame.groupby('funding_round_type')['raised_amount_usd'].mean().round(2).sort_values(
                  ascending=True).reset_index()
df_funding_type
```

```python
# Total amount invested in venture, funding round type

df_funding_type.loc[df_funding_type['funding_round_type'] == "venture"]
```

```python
# Total amount invested in angel funding round type

df_funding_type.loc[df_funding_type['funding_round_type'] == "angel"]
```

```python
# Total amount invested in seed funding round type

df_funding_type.loc[df_funding_type['funding_round_type'] == "seed"]
```

```python
# Total amount invested in private equity funding round type

df_funding_type.loc[df_funding_type['funding_round_type'] == "private_equity"]
```

```python
# Considering that Spark Funds wants to invest between 5 to 15 million USD per investment round, which investment type 
# is the most suitable for them?

df_funding_type.loc[(df_funding_type['raised_amount_usd'] >= 5.0) & (df_funding_type['raised_amount_usd'] <= 15.0)]
```

## Checkpoint 4: Sector Analysis 1

- Extract the primary sector of each category list from the category_list column
- Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that Others is 
  also considered one of the main sectors)
- Expected Results: Code for a merged data frame with each primary sector mapped to its main sector (the primary sector 
  should be present in a separate column).

```python
# Creating a new column with master_frame after splitting the category_list column based on '|' and adding the first split 
# result in a new column primary_sector.

master_frame['primary_sector'] = master_frame['category_list'].str.split('|', expand=True)[0]
# make it lower case as we would use for merging
master_frame['primary_sector'] = master_frame['primary_sector'].str.lower()
master_frame.head()
```

```python
# read mapping data file

mapping = pd.read_csv('D:\\Group-Assignments\\Course1\\mapping.csv', encoding='ISO-8859-1')
mapping.head()
```

```python
# inspect mapping data file

print("*********************************  Info *******************************************") 
print(mapping.info())
print("*********************************  Shape ******************************************") 
print(mapping.shape)
print("**************************** Columns having null values *****************************")
print(mapping.isnull().any())
print("*********************************  Description ************************************") 
mapping.describe()
```

```python
# create a sector map dataframe containing mapping between the column category_list and all the other columns from mapping
# dataframe using the melt function 
# This sector map dataframe will have only 2 columns : category_list and main_sector

df_sector_map = pd.melt(mapping, id_vars=["category_list"], var_name="main_sector", value_name="Value")

# drop the value column containing 0 and 1 values
df_sector_map.drop(df_sector_map[df_sector_map.Value == 0].index, inplace=True)
df_sector_map = df_sector_map.drop('Value',1)

# convert the values of category_list column to lower case as we would use it for merge later on with the master frame.
df_sector_map['category_list']= df_sector_map['category_list'].str.lower()

# rename the category_list to category_list_temp because, there is already a column in the master frame called category_list
df_sector_map = df_sector_map.rename(columns={'category_list': 'category_list_temp'})
df_sector_map.head(30)
```

```python
# Data Cleaning - The category_list_temp column of sector_df contains 53 items with 0 in place of na.

len(df_sector_map[df_sector_map['category_list_temp'].str.contains("0", na=False)])
```

```python
# Out of the above 53 rows, category_list_temp column also contains one value which is enterprise 2.0, in which 0 should 
# not be replaced with na

df_sector_map[df_sector_map['category_list_temp'].str.contains(".0", na=False, regex=False)]
```

```python
# Data Cleaning - Replace 0 with na for all 53 rows. Once that is done revert the na with 0 for the category with value 
# enterprise 2.0 

df_sector_map['category_list_temp'] = df_sector_map['category_list_temp'].replace(to_replace='0', value='na', regex=True)

# The above action also resulted in replacing enterprise 2.0 as enterprise 2.na , hence reverting the value to enterprise 2.0
df_sector_map['category_list_temp'] = df_sector_map['category_list_temp'].replace(to_replace='2.na', value='2.0', regex=True)

# Check if any value other than enterprise 2.0 contains 0 in category_list_temp column
df_sector_map.set_index('category_list_temp').filter(regex='0', axis=0)
```

```python
# Code for a merged data frame with each primary sector mapped to its main sector 
# (the primary sector should be present in a separate column).

master_frame = pd.merge(master_frame, df_sector_map, how = 'left', left_on = 'primary_sector', right_on = 'category_list_temp')
master_frame.head()
```

```python
# check if there are null values in any columns

round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# remove null rows from category_list_temp which would automatically drop the main_sector value named 'Blanks' as it is not 
# one of the main sectors

master_frame = master_frame.dropna(subset=['category_list_temp']) 
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
# drop column category_list_temp as it is not required for any more computation

master_frame = master_frame.drop('category_list_temp', 1)
master_frame.head()
```

## Checkpoint 5: Sector Analysis 2

`Now, the aim is to find out the most heavily invested main sectors in each of the three countries (for funding type FT and investments range of 5-15 M USD).`

`Create three separate data frames D1, D2 and D3 for each of the three countries containing the observations of funding type FT falling within the 5-15 million USD range. The three data frames should contain:`

- All the columns of the master_frame along with the primary sector and the main sector

- The total number (or count) of investments for each main sector in a separate column

- The total amount invested in each main sector in a separate column

`Using the three data frames, you can calculate the total number and amount of investments in each main sector.`

```python
# Top 3 countries =  USA, GBR and IND
# Funding Type based on the funding amount range between 5 to 15 million USD = venture

# For Country = USA, get all the columns of the master_frame along with the primary sector and the main sector

D1 = pd.DataFrame(master_frame.loc[(master_frame['country_code'] == 'USA') & (master_frame['funding_round_type'] == 'venture'
                            ) & (master_frame['raised_amount_usd'] >= 5.0) & (master_frame['raised_amount_usd'] <= 15.0)])
D1.head()
```

```python
# The total number (or count) of investments for each main sector in a separate column
# Note : The column 'sector_wise_invst_count' displays the total count of investments against each respective main sector.

D1['sector_wise_invst_count'] = D1.groupby('main_sector')['main_sector'].transform('count')
D1.head()
```

```python
# The total amount invested in each main sector in a separate column
# Note : The column 'sector_wise_tot_invst_amt' displays the total investment amount based on raised_amount_usd column
#        against each respective main sector.

D1['sector_wise_tot_invst_amt'] = D1.groupby('main_sector')['raised_amount_usd'].transform('sum').round(2)
D1.head()
```

```python
# Total number of Investments (count) in all 8 sectors

d1_tot_num_invst = D1.sort_values(['main_sector'],ascending=False).groupby('main_sector').head(1)
d1_total_investment_count = d1_tot_num_invst['sector_wise_invst_count'].sum()
print (d1_total_investment_count)
```

```python
# Total amount of Investments (USD) in all the 8 sectors
# The amount is in Million USD.

d1_tot_amount_invst = D1.sort_values(['main_sector'],ascending=False).groupby('main_sector').head(1)
d1_total_amount_invested = d1_tot_amount_invst['sector_wise_tot_invst_amt'].sum().round(2)
print (d1_total_amount_invested,"M USD")
```

```python
# Get the top 3 investment sectors based on the no. of investments
# Get the number of investments in the top 3 sectors

sorted_d1_on_num_invst = D1.sort_values(['sector_wise_invst_count'],ascending=False).groupby('sector_wise_invst_count').head(1)
sorted_d1_on_num_invst.head(3)
```

```python
# For top sector count-wise, which company received the highest investment?

D1.loc[D1['main_sector'] == 'Others', ['name', 'raised_amount_usd']].groupby('name').sum().round(2).sort_values(
                                                            by=['raised_amount_usd'], ascending = False).head(1)
```

```python
# For second best sector count-wise, which company received the highest investment?

D1.loc[D1['main_sector'] == 'Social, Finance, Analytics, Advertising', ['name', 'raised_amount_usd']].groupby('name'
                                                ).sum().round(2).sort_values(by=['raised_amount_usd'], ascending = False).head(1)
```
